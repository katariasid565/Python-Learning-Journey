```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

ðŸ”§ Step 1: Load and Prepare the Data

```python
df = sns.load_dataset("penguins")
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
X = df[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

ðŸ”® Step 2: Fit KMeans Clustering

```python
kmeans = KMeans(n_clusters = 3 , random_state= 42)
kmeans.fit(X_scaled)
```

```python
# Adding cluster labels
df['cluster'] = kmeans.labels_
```

```python
# ðŸ“ˆ Step 3: Visualize the Clusters

sns.scatterplot(data = df, x = 'flipper_length_mm', y = "bill_depth_mm", hue = "cluster", palette = "Set2")
plt.title("KMeans clusters of Penguins")
plt.show()
```

```python
### Customer Segmentation ####
```

```python
# Simulated customer data
data = {
    "CustomerID": range(1, 11),
    "Age": [19, 21, 20, 23, 31, 45, 40, 32, 28, 35],
    "Annual_Income": [15, 18, 20, 22, 50, 85, 70, 55, 48, 60],
    "Spending_Score": [39, 81, 6, 77, 40, 5, 10, 60, 48, 30]
}
```

```python
df3 = pd.DataFrame(data)
df3.head()
```

```python
features = ["Age", "Annual_Income", "Spending_Score"]
X = df3[features]
```

```python
scaler3 = StandardScaler()
```

```python
x_Scaled = scaler3.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
df3["Cluster"] = kmeans.fit_predict(x_Scaled)
```

```python
sns.scatterplot(x = "Annual_Income", y = "Spending_Score", data = df3, palette = "Set2", hue = "Cluster", s = 100)
plt.title("Customer Segmentation")
plt.xlabel("Annual Income {$}")
plt.ylabel("Spending_Score")
plt.show()
```
