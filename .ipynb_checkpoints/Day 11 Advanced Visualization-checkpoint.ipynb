ðŸ”¶ 1. Correlation Heatmap
Shows how features are related to each other.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

ðŸŒˆ 5. Styling Seaborn Plots

```python

sns.set(style = "whitegrid", palette= "muted")
```

```python


tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only = True)

sns.heatmap(corr, annot=True, cmap = 'coolwarm')
plt.title("Correlation Heatmap")

plt.show()
```

```python
titanic = sns.load_dataset("titanic")
corre = titanic.corr(numeric_only = True)

sns.heatmap(corre, annot=True, cmap= "coolwarm")
plt.title("Correlation HeatMap 2")
plt.show()
```

ðŸ”· 2. Box Plot
Helps to visualize distributions and detect outliers.

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.title("Box plot - Total Bill by Day")
plt.show()
```

```python
titanic.head()
```

```python
sns.boxplot(x = "sex", y = "age", data = titanic) ## x Should be categorical, y should be numerical.
plt.title("Box plot Class by Sex")
plt.show()
```

ðŸ”¶ 3. Violin Plot
Combines box plot and kernel density plot.

```python
sns.violinplot(x = 'day',y = 'total_bill', data = tips, inner="quartile", palette="pastel", hue = "day")
plt.title("Violin plot Total bills by Day.")
plt.show()
```

```python
sns.violinplot(x = "class", y = "age", data = titanic, inner = "quartile", palette = "pastel", hue = "class")
plt.title("Violin plot Class by age")
plt.show()
```

ðŸ”· 4. Pair Plot
Useful for visualizing pairwise relationships

```python
sns.pairplot(tips,hue = "sex")
plt.suptitle("Pairplot - Tips Dataset", y = 1)
plt.show()
```

```python
sns.pairplot(titanic, hue = "class")
plt.suptitle("pair plot Titanic Dataset", y = 1.1)
plt.show()
```
