```python
import matplotlib.pyplot as plt
import seaborn as sns
```

✅ Matplotlib – Basic plots (line, bar, scatter, histogram)
✅ Seaborn – Advanced plots (heatmaps, box plots, pair plots)

```python
# Sample Data
years = [2018, 2019, 2020, 2021, 2022]
sales = [5000, 7000, 8000, 6500, 9000]
```

```python
#Create Line plot
plt.plot(years, sales, marker = 'o', linestyle = '-', color = 'b', label = 'Sale Growth')

#Adding labels

plt.xlabel("Years")
plt.ylabel("Sales {$}")
plt.title("Sales Growth")
plt.legend()

##plot showing
plt.show()
```

```python
#Bar chart using matplotlib.

# Sample Data
categories = ["A", "B", "C", "D"]
values = [100, 200, 150, 250]

plt.bar(categories, values, color = ['red','blue','green','orange'])
plt.xlabel("Categories")
plt.ylabel("values")
plt.title("Bar Chart")
plt.show()
```

```python
## scatterplot using seaborn.
#load sample data
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.scatterplot(x = "total_bill", y="tip", data = tips, hue = "sex", style = "time")
plt.title("Total Bill vs Tip Amount")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
## Line plot using matplotlib.
plt.plot( marker = 'o', linestyle = "-", color = "y", label = 'Age-fare')

plt.xlabel("fare")
plt.ylabel("Age")
plt.title("Age-fare")
plt.legend()
plt.show()
```

```python
##Create Bar chart
sex = titanic["sex"]
survival = titanic["survived"]

plt.bar(sex, survival, color = ["red", "blue"])
plt.xlabel("sex")
plt.ylabel("Survivals")
plt.title("Survived class")
plt.show()
```
